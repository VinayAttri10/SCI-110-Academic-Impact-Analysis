```python
import pandas as pd
import random

num_rows = 15000
num_students = 1500 # Repeat IDs to build chronological history
student_ids = [random.randint(10000000, 99999999) for _ in range(num_students)]

# Data banks for generation
first_names = ['James', 'Mary', 'John', 'Patricia', 'Robert', 'Jennifer', 'Michael', 'Linda', 'William', 'Elizabeth', 'David', 'Barbara', 'Richard', 'Susan', 'Joseph', 'Jessica', 'Thomas', 'Sarah', 'Charles', 'Karen']
last_names = ['Smith', 'Johnson', 'Williams', 'Brown', 'Jones', 'Garcia', 'Miller', 'Davis', 'Rodriguez', 'Martinez', 'Hernandez', 'Lopez', 'Gonzalez', 'Wilson', 'Anderson', 'Thomas', 'Taylor', 'Moore', 'Jackson', 'Martin']
majors = ['Computer Science', 'Cybersecurity', 'Mathematics', 'Information Tech']
depts = ['SEC', 'CS', 'MATH', 'IT', 'ENG', 'PHY']
course_numbers = ['110', '101', '102', '201', '210', '305', '415', '499']
course_types = ['Lecture', 'Lab', 'Seminar', 'Online']
grades = ['A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D', 'F', 'W']
terms = ['FA21', 'SP22', 'SU22', 'FA22', 'SP23', 'SU23', 'FA23', 'SP24']

# --- NEW: Create consistent student profiles ---
# This ensures the same Student ID always has the same Name, Major, and PT/FT Status
student_profiles = {}
for sid in set(student_ids):
    student_profiles[sid] = {
        'First name': random.choice(first_names),
        'Last name': random.choice(last_names),
        'Major': random.choice(majors),
        'PT FT Status': random.choices(['Full-Time', 'Part-Time'], weights=[0.85, 0.15])[0] # 85% Full-Time
    }

data = []
print("Generating 15,000 rows with expanded columns...")

for _ in range(num_rows):
    student_id = random.choice(student_ids)
    profile = student_profiles[student_id] # Fetch the consistent student details
    
    # Force 'SEC 110' enough times so the dataframe isn't empty in your calculations
    is_target_course = random.random() < 0.15 
    if is_target_course:
        dept = 'SEC'
        course_num = '110'
    else:
        dept = random.choice(depts)
        course_num = random.choice(course_numbers)
        
    course_type = random.choice(course_types)
    
    # --- NEW: Logic for Credits and Enroll Status ---
    credits = 1 if course_type == 'Lab' else random.choice([3, 4])
    grade = random.choice(grades)
    enroll_status = "Withdrawn" if grade == 'W' else "Enrolled"

    # Append row exactly matching your new column structure
    data.append({
        'Student ID': student_id,
        'First name': profile['First name'],
        'Last name': profile['Last name'],
        'Major': profile['Major'],
        'Dept': dept,
        'Course Number': course_num,
        'Course Type': course_type,
        'Section': f"{random.randint(1, 15):02d}",
        'Credits': credits,
        'Grade': grade,
        'Enroll Status': enroll_status,
        'PT FT Status': profile['PT FT Status'],
        'Term': random.choice(terms)
    })

df = pd.DataFrame(data)

# Save to the exact path your Notebook uses
output_path = r'' # Update output directory path here
df.to_excel(output_path, index=False, sheet_name='All courses for SCI student')

print(f"Success! {len(df)} rows saved to {output_path}")
```
